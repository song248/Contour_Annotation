```python
import cv2, PIL
import numpy as np
import matplotlib.pyplot as plt
```

```python
def canny(img):
    g_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blr_img = cv2.GaussianBlur(g_img, (9, 9), 3.0)
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
    mph_img = cv2.morphologyEx(blr_img, cv2.MORPH_OPEN, kernel)
    edges = cv2.Canny(mph_img, 70, 130)
    edges_rgb = cv2.cvtColor(edges, cv2.COLOR_GRAY2RGB)
    return 255 - edges_rgb
```

```python
def img_vogi(img):
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
img = cv2.imread('ex_A.jpg', cv2.IMREAD_COLOR)
img_vogi(img)
```

```python
g_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blr_img = cv2.GaussianBlur(g_img, (9, 9), 3.0)
# blr_img = cv2.medianBlur(g_img, 5)
img_vogi(blr_img)
```

```python
# 모폴로지 열림 연산 적용 (작은 노이즈 제거)
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
mph_img = cv2.morphologyEx(blr_img, cv2.MORPH_OPEN, kernel)
img_vogi(mph_img)
```

```python
edges = cv2.Canny(mph_img, 70, 130)
img_vogi(edges)
```

```python
edges_rgb = cv2.cvtColor(edges, cv2.COLOR_GRAY2RGB)
img_vogi(edges_rgb)
```

```python
inv_edges_rgb = 255 - edges_rgb
img_vogi(inv_edges_rgb)
```

```python
plt.figure(figsize=(8, 6))

plt.subplot(2, 2, 1)
plt.title('Origin')
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title('GaussianBlur')
plt.imshow(blr_img, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title('Canny edge')
plt.imshow(edges)
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title('Final')
plt.imshow(inv_edges_rgb)
plt.axis('off')
```

```python
A_img = cv2.imread('ex_A.jpg', cv2.IMREAD_COLOR)
A_canny = canny(A_img)
B_img = cv2.imread('ex_B.jpg', cv2.IMREAD_COLOR)
B_canny = canny(B_img)
C_img = cv2.imread('ex_C.jpg', cv2.IMREAD_COLOR)
C_canny = canny(C_img)
D_img = cv2.imread('ex_D.jpg', cv2.IMREAD_COLOR)
D_canny = canny(D_img)
```

```python
plt.figure(figsize=(8, 6))

plt.subplot(2, 2, 1)
plt.title('A')
plt.imshow(A_canny)
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title('B')
plt.imshow(B_canny)
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title('C')
plt.imshow(C_canny)
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title('D')
plt.imshow(D_canny)
plt.axis('off')
```
