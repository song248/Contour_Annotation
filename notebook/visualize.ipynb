```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
txt_file_path = '111.txt'
image = cv2.imread('111.jpg')
image_height, image_width = image.shape[:2]
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Segmentation Visualization')
plt.axis('off')
plt.show()
```

```python
with open(txt_file_path, 'r') as file:
    lines = file.readlines()
```

```python
for line in lines:
    data = line.split()
    class_id = int(data[0])
    coordinates = list(map(float, data[1:]))
    
    # 정규화된 좌표를 실제 좌표로 변환
    points = np.array(coordinates).reshape(-1, 2)
    points[:, 0] *= image_width
    points[:, 1] *= image_height
    
    # 다각형 좌표를 정수로 변환
    points = points.astype(int)
    
    # 다각형 그리기
    cv2.polylines(image, [points], isClosed=True, color=(0, 255, 0), thickness=2)
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Segmentation Visualization')
plt.axis('off')
plt.show()
```
